## Puzzle 2

````python
with open('02/input.txt') as f:
  dims = f.readlines()
````

#### Part 1

Fortunately, every present is a box (a perfect right rectangular prism), which makes calculating the required wrapping paper for each gift a little easier: find the surface area of the box, which is 2*l*w + 2*w*h + 2*h*l. The elves also need a little extra paper for each present: the area of the smallest side.

````python
total = 0
for d in dims:
  (l, w, h) = (int(x) for x in d.strip().split('x'))
  
  area_1 = l*w
  area_2 = w*h
  area_3 = h*l

  total += 2*area_1 + 2*area_2 + 2*area_3 + min(area_1, area_2, area_3)

print(total)
````

### Part 2

The ribbon required to wrap a present is the shortest distance around its sides, or the smallest perimeter of any one face. Each present also requires a bow made out of ribbon as well; the feet of ribbon required for the perfect bow is equal to the cubic feet of volume of the present. Don't ask how they tie the bow, though; they'll never tell.

````python
total = 0
for d in dims:
  (l, w, h) = (int(x) for x in d.strip().split('x'))
  
  perim_1 = 2*l + 2*w
  perim_2 = 2*w + 2*h
  perim_3 = 2*h + 2*l
  vol = l*w*h

  total += min(perim_1, perim_2, perim_3) + vol

print(total)
````

### Part 1

A nice string is one with all of the following properties:

- It contains at least three vowels (aeiou only), like aei, xazegov, or aeiouaeiouaeiou.
- It contains at least one letter that appears twice in a row, like xx, abcdde (dd), or aabbccdd (aa, bb, cc, or dd).
- It does not contain the strings ab, cd, pq, or xy, even if they are part of one of the other requirements.

````python
def is_nice(s):
  return three_vowels(s) and double_letter(s) and bad_strings(s)

def three_vowels(s):
  total = 0
  for v in 'aeiou':
    total += s.count(v)
  if total > 2:
    return True
  else:
    return False

def double_letter(s):
  for idx in range(len(s) - 1):
    if s[idx] == s[idx + 1]:
      return True
  return False

def bad_strings(s):
  baddies = ['ab', 'cd', 'pq', 'xy']
  for b in baddies:
    if b in s:
      return False
  return True
````

````python
assert is_nice('ugknbfddgicrmopn')
assert is_nice('aaa')
assert not is_nice('jchzalrnumimnmhp')
assert not is_nice('haegwjzuvuyypxyu')
assert not is_nice('dvszwmarrgswjxmb')
````

````python
total = 0
for s in strings:
  if is_nice(s):
    total += 1
print(total)
````

### Part 2

Realizing the error of his ways, Santa has switched to a better model of determining whether a string is naughty or nice. None of the old rules apply, as they are all clearly ridiculous.

Now, a nice string is one with all of the following properties:

- It contains a pair of any two letters that appears at least twice in the string without overlapping, like xyxy (xy) or aabcdefgaa (aa), but not like aaa (aa, but it overlaps).
- It contains at least one letter which repeats with exactly one letter between them, like xyx, abcdefeghi (efe), or even aaa.

````python
def is_nice(s):
  return pair_repeat(s) and letter_between(s)

def pair_repeat(s):
  for idx in range(len(s) - 2):
    if s[idx:idx+2] in s[idx+2:]:
      # print(f'{s[idx:idx+2]}')
      return True
  return False

assert pair_repeat('xyxy')
assert pair_repeat('aabcdefgaa')
assert not pair_repeat('aaa')

def letter_between(s):
  for idx in range(len(s) - 2):
    if s[idx] == s[idx + 2]:
      return True
  return False

assert letter_between('xyx')
assert letter_between('abcdefeghi')
assert letter_between('aaa')

assert is_nice('qjhvhtzxzqqjkmpb')
assert is_nice('xxyxx')
assert not is_nice('uurcxstgmygtbstg')
assert not is_nice('ieodomkazucvgmuy')
````

## Puzzle 6

Because your neighbors keep defeating you in the holiday house decorating contest year after year, you've decided to deploy one million lights in a 1000x1000 grid.

Furthermore, because you've been especially nice this year, Santa has mailed you instructions on how to display the ideal lighting configuration.

Lights in your grid are numbered from 0 to 999 in each direction; the lights at each corner are at 0,0, 0,999, 999,999, and 999,0. The instructions include whether to turn on, turn off, or toggle various inclusive ranges given as coordinate pairs. Each coordinate pair represents opposite corners of a rectangle, inclusive; a coordinate pair like 0,0 through 2,2 therefore refers to 9 lights in a 3x3 square. The lights all start turned off.

To defeat your neighbors this year, all you have to do is set up your lights by doing the instructions Santa sent you in order.

For example:

- turn on 0,0 through 999,999 would turn on (or leave on) every light.
- toggle 0,0 through 999,0 would toggle the first line of 1000 lights, turning off the ones that were on, and turning on the ones that were off.
- turn off 499,499 through 500,500 would turn off (or leave off) the middle four lights.

````python
import numpy as np
grid = np.zeros((1000,1000), dtype=bool)

with open('06/input.txt') as f:
  inst = [l.strip().split() for l in f.readlines()]
for idx, i in enumerate(inst):
  if i[0] == 'turn':
    inst[idx] = i[1:]  
````

### Part 1

````python
def process_single_instruction(grid, line):
  command = line[0]
  x1, y1 = (int(j) for j in line[1].split(','))
  x2, y2 = (int(j) for j in line[3].split(','))

  if command == 'on':
    grid[x1:x2+1, y1:y2+1] = 1
  if command == 'off':
    grid[x1:x2+1, y1:y2+1] = 0
  if command == 'toggle':
    grid[x1:x2+1, y1:y2+1] = np.invert(grid[x1:x2+1, y1:y2+1])
  
  return grid

for l in inst:
  grid = process_single_instruction(grid, l)

print(grid.sum())
````

### Part 2

You just finish implementing your winning light pattern when you realize you mistranslated Santa's message from Ancient Nordic Elvish.

The light grid you bought actually has individual brightness controls; each light can have a brightness of zero or more. The lights all start at zero.

The phrase turn on actually means that you should increase the brightness of those lights by 1.

The phrase turn off actually means that you should decrease the brightness of those lights by 1, to a minimum of zero.

The phrase toggle actually means that you should increase the brightness of those lights by 2.

What is the total brightness of all lights combined after following Santa's instructions?

````python
grid = np.zeros((1000,1000), dtype=int)

def process_single_instruction(grid, line):
  command = line[0]
  x1, y1 = (int(j) for j in line[1].split(','))
  x2, y2 = (int(j) for j in line[3].split(','))

  if command == 'on':
    grid[x1:x2+1, y1:y2+1] = grid[x1:x2+1, y1:y2+1] + 1
  if command == 'off':
    grid[x1:x2+1, y1:y2+1] = grid[x1:x2+1, y1:y2+1] - 1
    grid = np.clip(grid, 0, np.inf)
  if command == 'toggle':
    grid[x1:x2+1, y1:y2+1] = grid[x1:x2+1, y1:y2+1] + 2
  
  return grid

for l in inst:
  grid = process_single_instruction(grid, l)

print(int(grid.sum()))
````

````python
# bitwise & with 16-bit integer to keep things as unsigned 16 bit ints:
#      & 0b1111111111111111
# i.e. -124 & 0b1111111111111111 == 65412

def proc_line(reg: Dict, l: str):
  inp, outp = (j.strip() for j in l.split('->'))
  inp = inp.split()

  # assignments
  if len(inp) == 1:
    if inp[0].isalpha() and inp[0] not in reg:
      # no-op because we don't have this value yet
      return None
    reg[outp] = reg[inp[0]] if inp[0].isalpha() else int(inp[0])

  # two parameter operations
  if len(inp) == 3:
    param_1, op, param_2 = inp

    if (param_1.isalpha() and param_1 not in reg) or \
       (param_2.isalpha() and param_2 not in reg):
       # no-op because we don't have the values we need yet
       return None
    
    param_1 = reg[param_1] if param_1.isalpha() else int(param_1) & 0b1111111111111111
    param_2 = reg[param_2] if param_2.isalpha() else int(param_2) & 0b1111111111111111

    if op == 'AND':
      reg[outp] = param_1 & param_2 & 0b1111111111111111
    elif op == 'OR':
      reg[outp] = param_1 | param_2 & 0b1111111111111111
    elif op == 'LSHIFT':
      reg[outp] = param_1 << param_2 & 0b1111111111111111
    elif op == 'RSHIFT':
      reg[outp] = param_1 >> param_2 & 0b1111111111111111

  # one parameter operations
  if len(inp) == 2:
    param = inp[1]

    if param.isalpha() and param not in reg:
      # no-op because we don't have the values we need yet
       return None
    
    param = reg[param] if param.isalpha() else int(param)

    reg[outp] = ~ param & 0b1111111111111111

  return reg

def proc_set(register: Dict, lines: List[str]):
  lines = lines[:]
  while len(lines) > 0:
    # print(register)
    for i, l in enumerate(lines):
      res = proc_line(register, l)
      if res is not None:
        register = res
        del lines[i]
  return register
````

Look-and-say sequences are generated iteratively, using the previous value as input for the next step. For each step, take the previous value, and replace each run of digits (like `111`) with the number of digits (`3`) followed by the digit itself (`1`).

For example:

- `1` becomes `11` (`1` copy of digit `1`).
- `11` becomes `21` (`2` copies of digit `1`).
- `21` becomes `1211` (one `2` followed by one `1`).
- `1211` becomes `111221` (one `1`, one `2`, and two `1`s).
- `111221` becomes `312211` (three `1`s, two `2`s, and one `1`).

Starting with the digits in your puzzle input, apply this process 40 times. What is the length of the result?

````python
def look_and_say(inp):
  idx = 0

  new_str = ''
  while idx < len(inp):
    # print(f'idx: {idx}')
    c = inp[idx]
    count = 1
    try: 
      while inp[idx] == inp[idx+1]:
        # print(f'idx: {idx}')
        count += 1
        idx += 1
    except Exception:
      pass
    
    idx += 1
    # print(f"idx: {idx}; {count} {c}s")
    new_str += str(count) + c

  return new_str
````

````python
# find cycles in the graph, then extract only complete cycles (i.e. the ones
# that have every node included)
cycles = sorted(list(nx.simple_cycles(G.to_directed())), key = lambda s: len(s)) 
cycles = [c for c in cycles if len(c) == len(G.nodes())]

lengths = {}
for c in cycles:
    c = c + [c[0]]   # complete the cycle
    lengths['-'.join(c)] = nx.path_weight(G, c, weight="weight")

print(max(lengths, key=lengths.get), lengths[max(lengths, key=lengths.get)])
````

````python
from tqdm import tqdm
with open('14/input.txt') as f:
    lines = [l.strip() for l in f.readlines()]

stats = {}
for ll in lines:
    l = ll.split(' ')
    name, speed, fly_time, rest_time = l[0], int(l[3]), int(l[6]), int(l[-2])
    stats[name] = {
        'speed': speed,
        'fly_time': fly_time,
        'rest_time': rest_time
    }

def distance_by_time(s: int, stats: Dict):
    i = 0
    dist = 0
    while i < s:
        if i < stats['fly_time']:
            dist += stats['speed']
            i += 1
        elif i < stats['fly_time'] + stats['rest_time']:
            dist += 0
            i += 1
        else:
            s -= (stats['fly_time'] + stats['rest_time'])
            i = 0
    
    return dist

def run_race(s: int):
    points = {
        'Rudolph': 0, 'Cupid': 0, 'Prancer': 0, 'Donner': 0, 'Dasher': 0, 
        'Comet': 0, 'Blitzen': 0, 'Vixen': 0, 'Dancer': 0
    }

    # process race at every second to see who's in the lead
    for i in tqdm(range(1, s+1)):
        dists = {}
        for deer in points.keys():
            dists[deer] = distance_by_time(i, stats[deer])
        leader = max(dists, key=dists.get)
        points[leader] += 1
    
    return points

run_race(2503)
````

````python
class Player:
    def __init__(self, name, hit_points, damage, armor):
        self.name = name
        self.hit_points = hit_points
        self.damage = damage
        self.armor = armor
        
    def __repr__(self):
        return f"{self.name}: H{self.hit_points} / D{self.damage} / A{self.armor}"
    
    def attack(self, other_player: Player, verbose=False):
        # Damage dealt by an attacker each turn is equal to the attacker's 
        # damage score minus the defender's armor score, 
        # An attacker always does at least `1` damage
        damage_dealt = max(self.damage - other_player.armor, 1)
        if verbose:
            print(f"{self.name} deals {self.damage} - {other_player.armor} = {damage_dealt} damage to {other_player.name}")
        other_player.hit_points -= damage_dealt

def play_game(player, boss, verbose=False):
    attacker = player
    defender = boss
    while player.hit_points > 0 and boss.hit_points > 0:
        attacker.attack(defender, verbose)
        if verbose:
            print(f"{player} / {boss}")
        if attacker == player:
            attacker = boss
            defender = player
        else:
            attacker = player
            defender = boss
    if player.hit_points > 0:
        return player
    else:
        return boss
            
````

````python
from collections import deque 
import random

class Player:
    def __init__(self, hit_points, mana):
        self.name = 'Player'
        self.hit_points = hit_points
        self.armor = 0
        self.mana = mana
        self.shield_turns = 0
        self.poison_turns = 0
        self.recharge_turns = 0
        self.mana_spent = 0
        self.spell_history = []
        
    def __repr__(self):
        return f"{self.name}: H-{self.hit_points} / M-{self.mana} / " + \
               f"A-{self.armor} / shield-{self.shield_turns} / " + \
               f"poison-{self.poison_turns} / recharge-{self.recharge_turns}"
    
    def magic_missle(self, other, verbose=False):
        self.mana -= 53
        self.mana_spent += 53
        other.hit_points -= 4
        if verbose:
            print(f"{self.name} casts magic_missle for 4 damage to {other.name} (cost 53 mana)")

    def drain(self, other, verbose=False):
        # Drain costs 73 mana. It instantly does 2 damage and heals you for 2 hit points.
        self.mana -= 73
        self.mana_spent += 73
        other.hit_points -= 2
        self.hit_points += 2
        if verbose:
            print(f"{self.name} casts drain for 2 damage to {other.name} and 2 hit points healing (cost 73 mana)")
    
    def shield(self, other, verbose=False):
        # Shield costs 113 mana. It starts an effect that lasts for 6 turns. 
        # While it is active, your armor is increased by 7.
        self.mana -= 113
        self.mana_spent += 113
        if self.shield_turns > 0:
            raise ValueError("cannot apply shield on top of existing shield")
        else:
            self.shield_turns = 6
    
    def poison(self, other, verbose=False):
        # Poison costs 173 mana. It starts an effect that lasts for 6 turns. 
        # At the start of each turn while it is active, it deals the boss 3 damage.
        self.mana -= 173
        self.mana_spent += 173
        if self.poison_turns > 0:
            raise ValueError("cannot apply poison on top of existing poison")
        else:
            self.poison_turns = 6
    
    def recharge(self, other, verbose=False):
        # Recharge costs 229 mana. It starts an effect that lasts for 5 turns. 
        # At the start of each turn while it is active, it gives you 101 new mana.
        self.mana -= 229
        self.mana_spent += 229
        if self.recharge_turns > 0:
            raise ValueError("cannot apply recharge on top of existing recharge")
        else:
            self.recharge_turns = 5
            
    def process_turn(self, other, spell, verbose=False):
        if verbose:
            if spell is None:
                print("-- Starting boss turn (player effects) --")
            else:
                print("-- Starting player turn --")
        spell_dict = {
            'magic_missile': self.magic_missle,
            'drain': self.drain,
            'shield': self.shield,
            'poison': self.poison,
            'recharge': self.recharge
        }
        # while shield is active, armor is increased by 7
        if self.shield_turns > 0:
            self.armor = 7
            self.shield_turns -= 1
            if verbose:
                print(f"shield was active, so armor set to 7; {self.shield_turns} turns remaining")
        else:
            self.armor = 0
        
        # while poison is active, does 3 damage
        if self.poison_turns > 0:
            other.hit_points -= 3
            self.poison_turns -= 1
            if verbose:
                print(f"poison was active, dealing 3 damage; {self.poison_turns} turns remaining")
        else:
            pass
        
        # while recharge is active, regenerate 101 mana
        if self.recharge_turns > 0:
            self.mana += 101
            self.recharge_turns -= 1
            if verbose:
                print(f"recharge was active, restoring 101 mana; {self.recharge_turns} turns remaining")
        else:
            pass
        
        # cast whatever spell (or pass, if it's not player's turn)
        if spell is None:
            pass
        else:
            if verbose:
                print(f"Casting {spell} spell")
            spell_dict[spell](other, verbose=verbose)
            self.spell_history.append(spell)
        
class Boss:
    def __init__(self, hit_points, damage):
        self.hit_points = hit_points
        self.damage = damage
        self.name = 'Boss'
        self.armor = 0
        self.turns = 0
        
    def __repr__(self):
        return f"{self.name}: H{self.hit_points} / D{self.damage} / A-{self.armor}"
    
    def process_turn(self, other: Player, verbose=False):
        # Damage dealt by an attacker each turn is equal to the attacker's 
        # damage score minus the defender's armor score, 
        # An attacker always does at least `1` damage
        if verbose:
            print("- Processing boss attack -")
        damage_dealt = max(self.damage - other.armor, 1)
        if verbose:
            print(f"{self.name} deals {self.damage} - {other.armor} = {damage_dealt} damage to {other.name}")
        other.hit_points -= damage_dealt

def play_game(player, boss, spells=None, verbose=False):
    current_turn = player
    while player.hit_points > 0 and boss.hit_points > 0 and player.mana > 0:
        # if verbose:
        #     print(f"{player} / {boss}")
            
        if current_turn == player:
            if spells is None:
                spell = input('spell to cast? ')
            elif spells == 'random':
                choices = ['magic_missile', 'drain']
                if player.shield_turns <= 1:
                    choices.append('shield')
                if player.poison_turns <= 1:
                    choices.append('poison')
                if player.recharge_turns <= 1:
                    choices.append('recharge')
                spell = random.choice(choices)
            else: 
                spell = spells.popleft()
            if check_status(player, boss, verbose) is not None:
                return check_status(player, boss)
            player.process_turn(other=boss, spell=spell, verbose=verbose)
            if check_status(player, boss, verbose) is not None:
                return check_status(player, boss)
            current_turn = boss
        else:
            if check_status(player, boss, verbose) is not None:
                return check_status(player, boss)
            player.process_turn(other=boss, spell=None, verbose=verbose)
            if check_status(player, boss, verbose) is not None:
                return check_status(player, boss)
            boss.process_turn(other=player, verbose=verbose)
            if check_status(player, boss, verbose) is not None:
                return check_status(player, boss)
            current_turn = player
        
        if verbose:
            print()
            
def check_status(player, boss, verbose=False):
    if verbose:
        print(f"- Player has {player.hit_points} HP, {player.armor} armor, {player.mana} mana")
        print(f"- Boss has {boss.hit_points} HP") 
    if player.hit_points <= 0:
        # print('player ran out of HP')
        return boss
    elif player.mana <= 0:
        # print('player ran out of mana')
        return boss
    elif boss.hit_points <= 0:
        # print('boss ran out of HP')
        return player
    else:
        return None
````
